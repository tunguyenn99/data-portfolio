```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

from tqdm.notebook import tqdm

from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

import time
import textwrap
```

```python
# Set up the Chrome webdriver
driver = webdriver.Chrome()
```

```python
url = "https://www.lazada.vn/routine-official/?from=wangpu&langFlag=vi&page=1&pageTypeId=2&q=All-Products"
driver.get(url)
```

```python
# Find all elements with data-qa-locator="product-item"
ele_indexes = driver.find_elements(By.XPATH, '//div[@data-qa-locator="product-item"]')

# Initialize lists to store the extracted data
data_listno = []
data_href = []
data_sold_qty = []
data_name_sku = []

# Iterate over each element to extract the required data
for ele_index in ele_indexes:
    # Extract data-listno
    data_listno.append(ele_index.get_attribute('data-listno'))
    
    # Extract href (only if it's not a duplicate)
    href_element = ele_index.find_element(By.XPATH, './/a[@age="0"]')
    href = href_element.get_attribute('href')
    if href not in data_href:
        data_href.append(href)
    
    # Extract "Đã bán" text
    try:
        sold_qty_element = ele_index.find_element(By.CSS_SELECTOR, 'span._1cEkb > span')
        sold_qty = sold_qty_element.text
    except:
        sold_qty = None
    data_sold_qty.append(sold_qty)

    # Extract Name
    name_element = ele_index.find_element(By.CSS_SELECTOR, 'div.RfADt > a')
    sku_name = name_element.get_attribute('title')
    # if sku_name not in data_name_sku:
    #     # data_href.append(href)
    data_name_sku.append(sku_name)

# Print the extracted data
for i in range(len(data_listno)):
    print("data-listno:", data_listno[i])
    print("href:", data_href[i])
    print("Đã bán:", data_sold_qty[i])
    print("Tên SKU:", data_name_sku[i])
    print('---')
```

```python
data = {
    'data-listno': data_listno,
    'href': data_href,
    'Đã bán': data_sold_qty,
    'SKU name': data_name_sku
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)
df
```
