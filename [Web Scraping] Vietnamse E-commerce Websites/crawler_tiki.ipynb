```python
### Mẹ và bé
all_hrefs = []

# Iterate through pages from 1 to 100
for page_num in tqdm(range(1, 51)):
    # Construct the URL for each page
    url = f"https://tiki.vn/dien-gia-dung/c1882?sort=top_seller&page={page_num}"
    
    # Open the URL in the browser
    driver.get(url)
    
    # Wait for 5 seconds to allow the page to load completely
    time.sleep(5)
    
    # Find all elements with href attribute
    href_elements = driver.find_elements(By.XPATH, "//a[@data-view-id='product_list_item']")
    
    hrefs = []

    # Extract href links from each element and append to the list
    for element in href_elements:
        href = element.get_attribute("href")
        all_hrefs.append(href)

    # all_hrefs.append(hrefs)

len(all_hrefs)
```

```python
df_href = pd.read_csv('href_diengiadung_tiki.csv')
all_hrefs = df_href['href'].to_list()
all_hrefs
```

```python
# all_hrefs[1940]
```

```python
brands = []
product_names = []
prices = []
discounts = []
cat_lv1s = []
cat_lv2s = []
cat_lv3s = []
review_counts = []
sold_quantities = []
hrefs_url = []
review_stars = []

for each_href in tqdm(all_hrefs):

    url = each_href
    driver.get(url)

    time.sleep(1)

    try:
        brand = driver.find_element(By.XPATH, '//a[@data-view-id="pdp_details_view_brand"]').text
    except:
        brand = None

    try:
        product_name = driver.find_element(By.XPATH, '//h1[@class="Title__TitledStyled-sc-c64ni5-0 iXccQY"]').text
    except:
        product_name = None

    try:
        price = driver.find_element(By.XPATH, '//div[@class="product-price__current-price"]').text.replace('₫','').replace('.','')
    except:
        price = None

    try:
        discount = driver.find_element(By.XPATH, '//div[@class="product-price__discount-rate"]').text.replace('-','')
    except:
        discount = None

    try:
        cat_lv1 = driver.find_element(By.XPATH, '//a[@class="breadcrumb-item" and @data-view-index="1"]').text
    except:
        cat_lv1 = None

    try:
        cat_lv2 = driver.find_element(By.XPATH, '//a[@class="breadcrumb-item" and @data-view-index="2"]').text
    except:
        cat_lv2 = None

    try:
        cat_lv3 = driver.find_element(By.XPATH, '//a[@class="breadcrumb-item" and @data-view-index="3"]').text
    except:
        cat_lv3 = None

    try:
        review_count = driver.find_element(By.XPATH, '//a[@data-view-id="pdp_main_view_review"]').text.replace('(','').replace(')','')
    except:
        review_count = None

    try:
        sold_quantity = driver.find_element(By.XPATH, '//div[@data-view-id="pdp_quantity_sold"]').text.replace('Đã bán ','')
    except:
        sold_quantity = None

    try: 
        review_star = driver.find_element(By.XPATH, '//div[@style="margin-right:4px;font-size:14px;line-height:150%;font-weight:500"]').text
    except:
        review_star = None
        
    brands.append(brand)
    product_names.append(product_name)
    prices.append(price)
    discounts.append(discount)
    cat_lv1s.append(cat_lv1)
    cat_lv2s.append(cat_lv2)
    cat_lv3s.append(cat_lv3)
    review_counts.append(review_count)
    sold_quantities.append(sold_quantity)
    hrefs_url.append(each_href)
    review_stars.append(review_star)
```

```python
data = {
    'Brand': brands,
    'Product Name': product_names,
    'Price': prices,
    'Discount': discounts,
    'Category Level 1': cat_lv1s,
    'Category Level 2': cat_lv2s,
    'Category Level 3': cat_lv3s,
    'Review Count': review_counts,
    'Sold Quantity': sold_quantities,
    'Review Star': review_stars,
    'URL': hrefs_url
}
df = pd.DataFrame(data)
df
```
