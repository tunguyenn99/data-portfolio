### Get all PAGE HREFs

```python
url_concung = 'https://concung.com/'
driver.get(url_concung)
```

```python
all_page_hrefs = []
all_page_titles = []

for i in range (0,11):
    all_page_href_ele = driver.find_elements(By.CSS_SELECTOR,f'#main-content > div > div.bg-whites > div > div.menu-style-scroll.pb-0.box-shadow-none.px-0 > div > ul > li:nth-child({i}) > a')

    # Iterate over found elements
    for element in all_page_href_ele:
        all_page_href = element.get_attribute("href")
        all_page_title = element.get_attribute("title")

        all_page_hrefs.append(all_page_href)
        all_page_titles.append(all_page_title)

data = {
    'page_href': all_page_hrefs
    ,'page_title':all_page_titles
}

df_all_page = pd.DataFrame(data)
df_all_page
```

```python
list_page_href = df_all_page['page_href'].to_list()
```

```python
len(list_page_href)
```

```python
hrefs = []
hrefs_solds = []

# Iterate through pages from 0 to 8
for page in list_page_href:
    url_page = page
    for page_num in range(1, 10):
        driver.get(f'{url_page}/?p={page_num}')

        href_ele = driver.find_elements(By.XPATH, "//a[@class='thumb-img box-shadow-none text-center d-inline-flex  w-100 text-center ' or @class='thumb-img box-shadow-none text-center d-inline-flex  text-center ']")
        href_sold_ele = driver.find_elements(By.XPATH, "//span[@class='selled' or @class='font-concung font-11']")

        # Check if lengths match
        if len(href_ele) != len(href_sold_ele):
            print(f"Lengths mismatch on page {url_page}, page number {page_num}")
            print(f"href_ele length: {len(href_ele)}, href_sold_ele length: {len(href_sold_ele)}")
        
        # Iterate over found elements
        for element in href_ele:
            try:
                href = element.get_attribute("href")
                hrefs.append(href)
            except Exception as e:
                print("Error getting href:", e)

        for element in href_sold_ele:
            try:
                hrefs_sold = element.text
                hrefs_solds.append(hrefs_sold)
            except Exception as e:
                print("Error getting sold quantity:", e)

data_hrefs_concung = {
    'href': hrefs,
    'Sold Quantity': hrefs_solds
}

df = pd.DataFrame(data_hrefs_concung)
print(df.head())
```

### LOOP ALL HREFS

```python
df = pd.read_csv('concung_hrefs_list.csv')

hrefs_sku_concung = df['href'].to_list()
```

```python
hrefs_sku_concung
```

```python
brands = []
product_names = []
prices = []
discounts = []
cat_lv1s = []
review_counts = []
sold_quantities = []
review_stars = []
hrefs_url = []

for each_href in tqdm(hrefs_sku_concung[356+49+1615+125:]):
    try:
        url = each_href
        driver.get(url)

        brand = driver.find_element(By.XPATH, '//a[@class="cc-pink-primary font-medium"]').text
    except Exception as e:
        brand = None
        print(f"Error finding brand for {url}: {e}")
    
    try:
        product_name = driver.find_element(By.XPATH, '//h1[@class=" product-name font-20 line-height-28 font-medium font-weight-normal mb-10"]').text
    except Exception as e:
        product_name = None
        print(f"Error finding product name for {url}: {e}")

    try:
        price = driver.find_element(By.XPATH, '//span[@class="product-price font-28 mr-2 font-medium order-price"]').text.replace('₫','').replace('.','')
    except Exception as e:
        price = None
        print(f"Error finding price for {url}: {e}")

    try:
        discount = driver.find_element(By.XPATH, '//span[@class="text-primary font-medium style-percent-product"]').text.replace('-','')
    except Exception as e:
        discount = None
        print(f"Error finding discount for {url}: {e}")

    try:
        cat_lv1 = driver.find_element(By.XPATH, '//*[@id="product"]/div[2]/div/div/div/div/ul/li[2]/a/span').text
    except Exception as e:
        cat_lv1 = None
        print(f"Error finding category level 1 for {url}: {e}")

    try:
        review_count = driver.find_element(By.CSS_SELECTOR, '#main-content > div > div.container.d-flex.p-20.bg-white.border-radius-12 > div.right-image-detail-top > div.product-meta.mb-3.pb-0.d-flex.justify-content-between > div.d-flex.align-center.font-16 > div > u:nth-child(3)').text
    except Exception as e:
        review_count = None
        print(f"Error finding review count for {url}: {e}")

    try:
        sold_quantity = driver.find_element(By.CSS_SELECTOR, '#main-content > div > div.container.d-flex.p-20.bg-white.border-radius-12 > div.right-image-detail-top > div.product-meta.mb-3.pb-0.d-flex.justify-content-between > div.d-flex.align-center.font-16 > div > u.font-bold.font-16.pl-15.ml-15').text
    except Exception as e:
        sold_quantity = None
        print(f"Error finding sold quantity for {url}: {e}")

    try:
        review_star = driver.find_element(By.CSS_SELECTOR, '#main-content > div > div.container.d-flex.p-20.bg-white.border-radius-12 > div.right-image-detail-top > div.product-meta.mb-3.pb-0.d-flex.justify-content-between > div.d-flex.align-center.font-16 > div > u:nth-child(1)').text
    except Exception as e:
        review_star = None
        print(f"Error finding review star for {url}: {e}")

    brands.append(brand)
    product_names.append(product_name)
    prices.append(price)
    discounts.append(discount)
    cat_lv1s.append(cat_lv1)
    review_counts.append(review_count)
    sold_quantities.append(sold_quantity)
    review_stars.append(review_star)
    hrefs_url.append(each_href)

data = {
    'Brand': brands,
    'Product Name': product_names,
    'Price': prices,
    'Discount': discounts,
    'Category Level 1': cat_lv1s,
    'Review Count': review_counts,
    'Sold Quantity': sold_quantities,
    'Review Star': review_stars,
    'URL': hrefs_url
}

df = pd.DataFrame(data)
df.head()
```
